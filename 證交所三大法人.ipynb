```python
import requests as r  # 引入 requests 庫，用於發送 HTTP 請求
from bs4 import BeautifulSoup as bs  # 引入 BeautifulSoup 庫，用於解析 HTML
import pandas as pd  # 引入 pandas 庫，用於資料處理

# 定義要爬取的網頁 URL
url = "https://www.twse.com.tw/fund/BFI82U?response=html"

# 送出 GET 請求，取得網頁內容
res = r.get(url)

# 使用 BeautifulSoup 解析網頁內容，並指定解析器為 'html5lib'
soup = bs(res.text, 'html5lib')

# 找到網頁中的表格標籤 <table>
tb = soup.find('table')

# 使用 pandas 將 HTML 表格轉換為 DataFrame
df = pd.read_html(str(tb))[0]

# 對表格中第二列開始的資料，應用格式化函數，將數字轉為帶有逗號的字串格式
df_comma = df.iloc[:, 1:].apply(lambda x: x.apply('{:,}'.format))

# 獲取表格的第一列資料
df.iloc[0:, :1]

# 將表格的第一列和格式化後的資料合併，形成新的 DataFrame
df_new = pd.concat([df.iloc[0:, :1], df_comma], axis=1)

# 顯示處理後的資料
df_new

```
