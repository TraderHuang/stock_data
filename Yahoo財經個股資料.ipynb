```python
#匯入 Yahoo 財經套件
import yfinance as yf
#匯入畫圖套件
import matplotlib.pyplot as plt
```

```python
#從Yahoo Finance
stock_NVDA = yf.download("NVDA","2024-08-01","2024-08-16")
stock_NVDA.head()
```

```python
#將股價資料存成CSV檔案
stock_NVDA.to_csv("stock_NVDA.csv")
```

```python
#繪製股價趨勢圖
fig = plt.figure(figsize = (15, 5))
plt.plot(stock_NVDA["Close"],color = "red")
plt.title("NVDA Close price")
plt.show()
```
